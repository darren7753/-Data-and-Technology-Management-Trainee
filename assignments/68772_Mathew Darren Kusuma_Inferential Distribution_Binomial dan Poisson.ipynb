# Binomial Distribution

## What It Is

The **binomial distribution** is a discrete probability distribution that models the number of successes in a fixed number of independent experiments or trials, each with the same probability of success. Here, a "success" is defined as the outcome of interest (e.g., getting heads when flipping a coin or making a sale on a customer call).

## What It's Used For

The binomial distribution is commonly used in scenarios where:

- There are a fixed number of trials ($ n $).
- Each trial has only two possible outcomes: success or failure.
- The probability of success ($ p $) is the same in each trial.
- The trials are independent of one another.

## Formula for Binomial Distribution

The probability of getting exactly $ k $ successes in $ n $ trials is given by the formula:

$$ P(X = k) = \binom{n}{k} p^k (1 - p)^{n - k} $$

where:
- $ P(X = k) $ = probability of getting exactly $ k $ successes.
- $ n $ = total number of trials.
- $ k $ = number of successes.
- $ p $ = probability of success in a single trial.
- $ (1 - p) $ = probability of failure in a single trial.
- $ \binom{n}{k} $ = "n choose k", the binomial coefficient, calculated as:

$$ \binom{n}{k} = \frac{n!}{k! (n - k)!} $$

where $ n! $ (n factorial) represents the product of all positive integers from 1 to $ n $.

## Real-Life Example

A company wants to know the probability of getting exactly 5 sales from 10 customer calls, assuming a 30% success rate per call.

Here:
- $ n = 10 $ (number of calls)
- $ p = 0.3 $ (probability of success per call)
- $ k = 5 $ (number of desired successes)  

### Manual Calculation

$$ P(X = 5) = \binom{10}{5} \cdot p^5 \cdot (1 - p)^{10 - 5} $$

$$ P(X = 5) = \frac{10!}{5! \cdot (10 - 5)!}  \cdot (0.3)^5 \cdot (0.7)^5 $$

$$ P(X = 5) \approx 10.3\% $$

So, the probability of getting exactly 5 sales out of 10 calls, given a 30% success rate, is approximately **10.3%**.

### Python Calculation

```python
n = 10
p = 0.3
x = np.arange(0, n + 1)
y = binom.pmf(x, n, p)
blue = "#123b67"
grey = "#a7a7a8"
colors = [blue if val == 5 else grey for val in x]

plt.figure(figsize=(10, 5), dpi=200)
plt.bar(x, y, color=colors)
plt.title("Binomial Distribution", fontsize=14)
plt.xlabel("Number of Sales (Successes)", fontsize=12)
plt.ylabel("Probability", fontsize=12)
plt.xticks(x)

plt.annotate(
    f"Assuming a 30% success rate per call,\nthe probability of getting exactly 5 sales is {y[5] * 100:.1f}%",
    xy=(5, y[5]), 
    xytext=(7.5, y[5] + 0.05),
    ha="center",
    va="bottom",
    fontsize=10,
    color="white",
    bbox=dict(
        boxstyle="round,pad=0.5",
        fc=blue,
        ec="none"
    ),
    arrowprops=dict(
        arrowstyle="->",
        color=blue,
        linewidth=1.5
    )
)

for spine in ["top", "right", "left", "bottom"]:
    plt.gca().spines[spine].set_visible(False)

plt.show()
```

The graph visualizes the binomial distribution for this example. The bar at 5 indicates the probability of obtaining exactly 5 sales, which is highlighted in blue. As the probability is highest around 3, this aligns with the expectation of around 3 successes, given the 30% success rate per call. The further the number of successes deviates from this value, the lower the probability, demonstrating the distribution's characteristic bell shape.

# Poisson Distribution

## What It Is

The **Poisson distribution** is a discrete probability distribution that expresses the probability of a given number of events occurring in a fixed interval of time or space, provided these events happen with a known constant mean rate ($ \lambda $) and independently of each other.

## What It's Used For

The Poisson distribution is commonly used in scenarios where:

- We count the number of occurrences of an event over a fixed period or space.
- The events occur independently.
- The average number of occurrences ($ \lambda $) is known.
- The number of events in one interval is independent of the number of events in other intervals.

## Formula for Poisson Distribution

The probability of observing $ k $ events in a fixed interval is given by the formula:

$$ P(X = k) = \frac{\lambda^k e^{-\lambda}}{k!} $$

where:
- $ P(X = k) $ = probability of $ k $ events occurring.
- $ \lambda $ = average number of events per interval.
- $ k $ = number of occurrences.
- $ e $ = Euler's number (approximately 2.718).

## Real-Life Example

A call center receives an average of 5 calls per hour. What is the probability of receiving exactly 3 calls in a given hour?

Here:
- $ \lambda = 5 $ (average rate of occurrence)
- $ k = 3 $ (number of occurrences)

### Manual Calculation

$$ P(X = 3) = \frac{5^3 e^{-5}}{3!} $$

$$ P(X = 3) = \frac{125 \times 0.006738}{6} $$

$$ P(X = 3) \approx 14\% $$

So, the probability of receiving exactly 3 calls in an hour is approximately **14%**.

### Python Calculation

```python
mu = 5
x = np.arange(0, 15)
y = poisson.pmf(x, mu)
blue = "#123b67"
grey = "#a7a7a8"
colors = [blue if val == 3 else grey for val in x]

plt.figure(figsize=(10, 5), dpi=200)
plt.bar(x, y, color=colors)
plt.title("Poisson Distribution", fontsize=14)
plt.xlabel("Number of Calls", fontsize=12)
plt.ylabel("Probability", fontsize=12)
plt.xticks(x)

plt.annotate(
    f"Assuming an average of 5 calls per hour,\nthe probability of receiving exactly 3 calls is {y[3] * 100:.1f}%",
    xy=(3, y[3]),
    xytext=(11, y[3] + 0.02),
    ha="center",
    va="bottom",
    fontsize=10,
    color="white",
    bbox=dict(
        boxstyle="round,pad=0.5",
        fc=blue,
        ec="none"
    ),
    arrowprops=dict(
        arrowstyle="->",
        connectionstyle="angle,angleA=0,angleB=90,rad=1",
        color=blue,
        linewidth=1.5
    )
)

for spine in ["top", "right", "left", "bottom"]:
    plt.gca().spines[spine].set_visible(False)

plt.show()
```

The graph visualizes the Poisson distribution for this example, with the bar at 3 calls highlighted in blue. This bar shows the probability of exactly 3 calls occurring within an hour, which is 14%. The distribution demonstrates that while it's most likely to receive around 5 calls (the average rate), there is still a significant probability of receiving fewer calls, such as 3. The probabilities decrease rapidly as the number of calls deviates further from the average.

# Conclusion

- **Binomial distribution** is used for modeling the number of successes in a fixed number of independent trials, with the same probability of success in each trial.
- **Poisson distribution** is used for modeling the number of events happening in a fixed interval of time or space, with a known constant average rate.
